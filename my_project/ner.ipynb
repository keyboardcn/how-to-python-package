```python
import spacy
# Load a pre-trained model
# You can use 'en_core_web_sm', 'en_core_web_md', or 'en_core_web_lg'
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Apple is looking at buying U.K. startup for $1 billion. Tim Cook (timcook@gmail.com) visited London last week."

# Process the text
doc = nlp(text)

# Iterate over the named entities
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")

# You can also visualize the entities (especially in a Jupyter Notebook)
# from spacy import displacy
# displacy.render(doc, style="ent", jupyter=True)
pattern =r"([a-zA-Z0-9]\w+@[a-zA-Z0-9]+\.[a-z]{2,4})"
```
